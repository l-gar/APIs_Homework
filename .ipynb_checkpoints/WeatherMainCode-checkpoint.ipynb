### WeatherPy

## Analysis
* As latitude increases temperature decreases in this data set. However, the data shows a bell curve meaning that with more data we would see the same happening at the lowest latitudes. 
* Overall latitudes does not seem to correlate well with either cloudiness or humidity. However, most data seems to fall on the higher end of humidity while cloudiness does not appear to have a discernable pattern. 
* Latitude does not seem to be a factor in windspeed. 


```python
import pandas as pd
import matplotlib.pyplot as plt
import openweathermapy as owm
import csv
import numpy as np
import requests
import time
from citipy import citipy
from pprint import pprint
from api_keys import api_key
```

```python
# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

### Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
print(cities)
```

### Perform API Calls

```python
settings = {"units": "metric", "appid": api_key}

weatherc_df = pd.DataFrame( columns = ["City", "Cloudiness", "Country", "Date", "Humidity", "Lat", "Lng", "Max Temp", "Wind Speed"])

current_weather_paris = owm.get_current("Paris", **settings)
print(current_weather_paris)
call_count = 1
sets = 1
index = 0
```

```python
print(f'Beggining Data Retrieval \n---------------------------------------')
for city in cities:
    try: 
        print(f'Processing Record {call_count} of set{sets} | {city}')
        city_data = owm.get_current(city, **settings)
        weatherc_df.set_value(index, "City", city_data["name"])
        weatherc_df.set_value(index, "Cloudiness", city_data("clouds.all"))
        weatherc_df.set_value(index, "Country", city_data("sys.country"))
        weatherc_df.set_value(index, "Date", city_data("dt"))
        weatherc_df.set_value(index, "Humidity", city_data("main.humidity"))
        weatherc_df.set_value(index, "Lat", city_data("coord.lat"))
        weatherc_df.set_value(index, "Lng", city_data("coord.lon"))
        weatherc_df.set_value(index, "Max Temp", city_data("main.temp_max"))
        weatherc_df.set_value(index, "Wind Speed", city_data("wind.speed"))
        index = index + 1
    except Exception as e:
        print(type(e))
        print(str(e))
    call_count = call_count + 1
    if call_count == 51:
        call_count = 1
        sets = sets + 1
        time.sleep(60)
        
print(f'------------------------------ \n End of Data Retrieval \n------------------------------')
```

## Latitude vs. Temperature Plot

```python
plt.scatter(x, y, c= y ,cmap = 'coolwarm', edgecolor = 'black', alpha=.9, linewidths=.4)
plt.title('City Latitude vs. Max Temperature (08/22/2018)')
plt.xlabel('Latitude')
plt.ylabel('Max Temperature (F)')
#plt.ylim(top = y.max()+10)
plt.legend(prop={'size': 10}, markerscale = .75)
plt.grid(bool)
plt.tight_layout()

plt.savefig("Weatherpy_LatitudeTemperature_fig")

plt.show()
```
